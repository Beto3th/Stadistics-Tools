```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('houses.csv')
```

```python
datos.head(5)
```

```python
datos.info()
```

```python
# Caracteristicas que el algoritmo va a aprendizaje
X = datos[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y = datos['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Crear el objeto de la regresion lineal 
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train) # Entrenamiento de la regresión 
```

```python
y_pred = modelo.predict(X_test)
```

```python
plt.scatter(y_test, y_pred)
# plt.plot(X_test['Avg. Area Income'], y_pred)
```
