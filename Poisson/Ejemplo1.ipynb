### Probabilidad de Clientes en un banco
Utilizaremos la distribucion de Poisson para simular a llegada de clientes a un banco durante un periodo de 8hrs, la taza media de llegada de clientes se calcula dividiendo el npumero total de clientes entre el tiempo total. Luego, generamos los tiempos de llegada de los clientes utilizando la distribucion de Poissson y visualizamos el proceso de llegada 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
tiempor_total = 8*60 # 8 horas en minutos
clientes = 100 # Numero total de clientes

# calcular la tasa media de llegada de clientes 
tasa = clientes / tiempor_total

# Generar os tiempo de llegadas de los clientes utilizando la distribucion de Poisson 
tiempos_llegada = np.random.poisson(tasa, clientes)
tiempos_llegada = np.cumsum(tiempos_llegada)

# Grafica de llegada de clientes durante el tiempo total
plt.plot(tiempos_llegada, np.arange(1, clientes+1), marker = 'o', linestyle='-', color='y')
plt.xlabel("Tiempor (minutos)")
plt.ylabel("Numero de clientes")
plt.title("Llegada de clientes a un banco en 8 hrs")
plt.show()
```

## Ocurrencia de llamadas telefónicas 
Vamos a modelar la ocurrencia de llamadas telefonicas ul¿tilizando la distribucion de Poisson.
Calcular la tasa media de llegada de llamadas dividiendo el num total de llamadas entre el tiempo total de observacion. Luego, generamots los tiempos de ocurrencia de llamadas utiklizando la distr de Poisson y visualizamos en una grafixa el proceso a lo largo del tiempo.

```python
# Parametros 
tiempo_total = 8 * 60 * 60 #Tiempo en seg
llamadas = 200 # llamadas totales 
# Calcular la tasa media de llegada de llamadas 
tasa = llamadas / tiempo_total

# Calcular los tiempos de ocurrencia de llamadas utilizando la distrb de Poisson 
tiempos_llamadas = np.random.poisson(tasa, llamadas)
tiempos_llamadas=np.cumsum(tiempos_llamadas)

# Grafica de ocurrencia de llamadas
plt.step(tiempos_llamadas, np.arange(1,llamadas+1), where='post')
plt.xlabel("Tiempo en segundos")
plt.ylabel("Numero de llamadas")
plt.title("Ocurrencia de llamadas ")
plt.show()
```
