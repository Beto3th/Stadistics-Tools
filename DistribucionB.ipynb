## Distribucion binomial

### Lanzamiento de una moneda


```python
#CÃ¡lculos numericos 
import numpy as np 
# Graficar 
import matplotlib.pyplot as plt
```

```python
# Definicion de PArametros 
n = 5 # Numero de ensayos
p = 0.5 # Probabilidad de caso de exito
```

```python
# Generacion de valores para la distribucion 
x = np.arange(0, n+1) # Valores Posibles de exito (0, n)
probabilidades = np.zeros(n+1)
```

```python
#Calcular la probabilidad para cada valor de x
for i in range(n+1):
    probabilidades[i] = np.math.comb(n, i) * (p**i)  * ((1-p)**(n-i))
```

```python
# Graficar la distribucion Binomial 
# Asegurarse que x y probabilidades sean del mismo tamaÃ±o 
plt.bar(x, probabilidades)
plt.title("Distribucion Binomial de lanzamiento de moneda")
plt.xlabel("NÃºmero de Ã©xitos")
plt.ylabel("Probabilidad")
```

Esta grafica muestra probabilidades de Ã©xito en 5 lanzamientos de moneda, y se puede observar que la distribucion de los eventos se asemeja a una distribucion normal debido a las probabilidades de Ã©xito.
