### 1.2. Generación de Datos
Para realizar esta práctica es necesario generar un conjunto de datos simulados de **un millón** de estudiantes con sus alturas. Para ello, utilizaremos técnicas de simulación y la distribución normal para crear un conjunto de datos que represente una muestra de estudiantes

```python
np.random.seed()
datos = np.random.normal(loc=150,scale=5, size=1000000 ) # Generar las alturas
np.round(datos)
```

### 1.3. Cálculo de Medidas de Tendencia Central:
- Calcula la media, la mediana y la moda de este conjunto de datos.

```python
datos
```

```python
#datos_r=[round(x) for x in datos] # Redondear los datos de alturas
```

```python
#datos_r
```

```python
media = sum(datos)/len(datos) # Calculo de media 
mediana = np.median(datos) # Calculo de mediana

frecuencias = {} # Diccionario
for i in datos:
    frecuencias.setdefault(i, 0)
    frecuencias[i] += 1

moda = max(frecuencias, key=frecuencias.get)

print(f"La media es : {media}")
print(f"La mediana es : {mediana}")
print(f"La moda es : {moda}")
```

### 1.4. Cálculo de Medidas de Dispersión:
- Siguiendo con los datos de altura de estudiantes, calcula la desviación estándar y el rango intercuartílico.
### 1.5. Gráficos Descriptivos:
- Grafica un histograma para representar la distribución de las alturas.
- Además del histograma, construye un diagrama de caja (boxplot) para visualizar la dispersión de los datos.

```python
# Desviacion estandar 
desviacion_estandar = np.std(datos)
print(f"La desviacion estandar es: {desviacion_estandar}")

# Rango intercuartílico
Q1= np.percentile(datos, 25)
Q3= np.percentile(datos, 75)

RI = Q3 - Q1

print(f"El rango intercuartílico es: {RI}")
```

```python
# Histograma
plt.figure(figsize=(12, 8))
plt.hist(datos, bins=50)
plt.title("HISTOGRAMA DE ALTURAS")
plt.xlabel("ALTURAS")
plt.ylabel("Frecuencias")
plt.show()
```

```python
# BOXPLOT
plt.figure(figsize=(20,15))
plt.boxplot(datos)
plt.title("Diagrama de Caja")
plt.ylabel("Valor del dato")
plt.show()
```

### 2.1 Teoría de la Distribución Normal:
- Describe brevemente qué es la distribución normal y menciona algunas de sus propiedades importantes.

#### DISTRIBUCIÓN NORMAL
- La distribución normal es una distribución teórica de los valores. 

#### PROPIEDADES IMPORTANTES
- A menudo recibe el nombre de curva de campana porque la representación visual de esta distribución tiene forma de campana. 
- Es teórico porque su distribución de frecuencias se deriva de una fórmula en lugar de la observación de los datos reales.
- Las distribuciones de muchos campos en el mundo real se parecen a la distribución normal.
- La distribución normal es simétrica acerca de la media.

### 2.2 Aplicación de la Distribución Normal:
- Supongamos que las alturas de los estudiantes de la universidad siguen una distribución normal con una media de 170 cm y una desviación estándar de 5 cm. Calcula la probabilidad de que un estudiante elegido al azar tenga una altura menor a 175 cm.
- Calcula también la probabilidad de que un estudiante tenga una altura entre 165 cm y 180 cm.

```python
alturas = np.random.normal(loc=170, scale=5, size=1000000)
```

```python
m = sum(1 for x in alturas if x < 175)
prob = m / len(alturas)

print(f"La probabilidad de de seleccionar a alguien menor a 175cm es: {prob}")
```

```python
m = sum(1 for x in alturas if x < 180 and x > 165)
prob = m / len(alturas)

print(f"La probabilidad de de seleccionar a alguien menor a 175cm es: {prob}")
```
