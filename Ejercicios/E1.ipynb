# Ejercicio: Funciones Lineales

```python
import matplotlib.pyplot as plt
import numpy as np
```

## f(x) = 2x+3  Arreglo aleatrio de 10 n√∫meros enteros entre 0 y 100

```python
np.random.seed()

x1 = np.random.randint(0, 100, 10)

print(x1)

y1 = (2 * x1) + 3

plt.plot(x1, y1)
plt.title("f(x) = 2x+3")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x2 = np.random.uniform(-10, 10, 20)

print(x2)

y2 = (0.5 * x2) - 1

plt.plot(x2, y2)
plt.title("f(x) = 0.5x-1")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
