```python
# Importa librerias
import pandas as pd # Manejar csv (comma separated values)
import numpy as np  # Manejar valores numericos 
import matplotlib.pyplot as plt # Graficas 
from sklearn.model_selection import train_test_split # Sickit-learn--Herramienta par ciencia de datos
from sklearn.linear_model import LinearRegression # Clase para regresion lineal 
```

```python
datos = pd.read_csv("salary.csv.xls")
```

```python
datos.head(5)
```

```python
plt.bar(datos["YearsExperience"], datos["Salary"])
```

```python
x = datos.iloc[:, :1] # Datos que se va a aprender
y = datos.iloc[:, 1:] # Etiquetas que le dara a los datos:
```

```python
# x_train <- valores de entrenamiento 
# y_train <.- Etiquetas de entrenamiento
# x_test <- Valores para probar el modelo 
# y_test <- Etiquetas para comprobrar el modelo 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Crear la regresion Lineal
regresion = LinearRegression()
# Entrenar el modelo
regresion.fit(x_train, y_train)
```

```python
y_pred=regresion.predict(x_test) # calcular las predicciones usando el modelo
```

```python
# visualizar la regresion lineal
plt.scatter(x_test, y_test) 
plt.plot(x_test, y_pred)
plt.title("Salario Vs Años de Experiencia")
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario")
plt.show()
```
