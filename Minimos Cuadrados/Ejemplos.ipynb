```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f1(x):
    return 1+2*x
```

```python
def f2(x):
    return 0.368 +2.206*x
```

```python
x = np.linspace(0, 100, 100 )
```

```python
y1 = f1(x)
```

```python
y2 = f2(x)
```

```python
plt.plot(x, y1, label="y=1+2x")
plt.plot(x, y2, label= "y=0.368+2.206x")
plt.title("Graficos de minimos cuadrados")
plt.grid()
plt.show()
```
