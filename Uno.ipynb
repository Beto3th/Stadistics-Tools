```python
import numpy as np
```

Generar datos aleatoreos

```python
np.random.seed()
# loc = media
# scale = incremento
# size = cantidad n√∫meros
datos = np.random.random(1000)
```

```python
datos
```

```python
sum(datos)/len(datos)
```

```python
np.mean(datos)
```

Mediana


```python
datos_ordenados = sorted(datos)
n = len(datos_ordenados)
if n % 2 == 0:
    mediana = (datos_ordenados[n//2 - 1] + datos_ordenados[n//2]/2)
else:
    mediana = datos_ordenados[n//2]
```

```python
mediana
```

```python
np.median(datos)
```

MODA

```python
frecuencias = {} # Diccionario
for i in datos:
    frecuencias.setdefault(i, 0)
    frecuencias[i] += 1
```

```python
frecuencias
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(12, 8))
plt.hist(datos, bins=50)
plt.title("HISTOGRAMA DE FRECUENCIAS")
plt.xlabel("Valores")
plt.ylabel("Frecuencias")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(datos)
plt.title("Diagrama de Caja")
plt.ylabel("Valor del dato")
plt.show()
```
