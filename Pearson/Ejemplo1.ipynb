```python
import math
```

```python
# Generar aleatoriamente los valores de las variables x, y
x = [ 8, 1,2.5,4,7,9.3,1.3,6.1]
y = [9.8,2,4.8,2.1,7,3.2,5.4, 0]
```

```python
x = [ 1.1, 3.4,5.2,7.8,8.0,9.7,3.3,1.0]
y = [ 1.0, 3.6,5.1,7.6,8.0,9.7,3.2,1.1]
```

```python
x_media = sum(x) / len(x)
y_media = sum(y) / len(y)
```

```python
numerador = sum((xi - x_media)  * (yi - y_media) for xi, yi in zip(x,y))
```

```python
den = math.sqrt(sum((xi - x_media)**2 for xi in x)) * math.sqrt(sum((yi - y_media)**2 for yi in y))
```

```python
r = numerador / den 
```

```python
print(f"El coeficiente de Pearson es {r}")
```

```python
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
dataset = load_breast_cancer()
```

```python
X = dataset.data
y = dataset.target
```

```python
X.shape
```

```python
def obtener_coeficiente_pearson(X):
    # Matriz de correlaci√≥n de n x n 
    n = X.shape[1]
    matriz = np.empty((n, n))
    for x in range (X.shape[1]):
        for y in range(X.shape[1]):
            X_media = sum(X [:, x]) / X.shape[0]
            y_media = sum(X[:, y]) / X.shape[0]

            num = sum((xi - x_media) * (yi - y_media) for xi, yi in zip(X[: , x], X[: , y])) 
            den = math.sqrt(sum((xi - X_media) ** 2 for xi in X[:, x])) * math.sqrt(sum((yi - y_media) ** 2 for yi in X[:, y]))
            r = num / den  
            matriz[x, y] = r 

    return matriz 
```

```python
matriz_correlacion = obtener_coeficiente_pearson(X)
```
